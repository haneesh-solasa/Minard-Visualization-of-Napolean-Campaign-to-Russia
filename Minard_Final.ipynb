```python
import pandas as pd
import altair as alt
red="#FF0000"
green="#006400"
black="#000000"
```

```python
#Cities Read
data1=pd.read_excel('cities.xlsx')
Cities_Napolean1 = pd.concat([data1['LONC'],data1['LATC'],data1['CITY']],axis=1)
Cities_Napolean1 = Cities_Napolean1.dropna()
#K Val Read
datan=pd.read_excel('K_Army.xlsx')
K_Napolean1 = pd.concat([datan['LONT'],datan['LATT'],datan['K']],axis=1)
#Temperature Read
data2=pd.read_excel('Temperature.xlsx')
Temperature = pd.concat([data2['LONT'],data2['TEMP'],data2['DAYS'],data2['Date'],data2['Display']],axis=1)
Temperature = Temperature.dropna(how='all')
Temperature = Temperature.fillna("")
#Army Read
data3 = pd.read_excel('army_final.xlsx')
Army=pd.concat([data3['LONP'],data3['LATP'],data3['LONT'],data3['LATT'],data3['SURV'],data3['DIR'],data3['DIV']],axis=1)
```

```python
#City Names
city_chart = alt.Chart(Cities_Napolean1).mark_text(font='Monospace',
    fontSize=13, dy=25,dx=-10 
).encode(longitude='LONC',latitude='LATC',text='CITY', 
).project(type="mercator")
#City Points
city_chart2 = alt.Chart(Cities_Napolean1).mark_circle( dx=0,color='black').encode(
    longitude='LONC',latitude='LATC',text='CITY',).project(type="mercator")
#K Value
Final_K = alt.Chart(K_Napolean1).mark_text(font='Monospace',
    fontSize=13, dy=30,dx=40,color='indigo'
).encode(longitude='LONT',latitude='LATT',text='K',
).project(type="mercator")
```

```python
Final_Cities=city_chart+city_chart2+Final_K
Final_Cities = Final_Cities.properties(height=200,width=1200)
```

```python
#Creating X & Y Axis
X_axis=alt.X('LONT',scale=alt.Scale(domain=[Cities_Napolean1["LONC"].min(), Cities_Napolean1["LONC"].max()]))
Y_axis= alt.Y('TEMP',axis=alt.Axis(title="Temperature",grid=True,orient='left')) 

#Plotting the Temp Graph on Chart
Chart_Temp= alt.Chart(Temperature).mark_line(color='black').encode(
    x=X_axis,
    y=Y_axis
    )
#Plotting the Temperature Values on CHart
Temperature_Graph=alt.Chart(Temperature).mark_text(dx=-10,dy=-20,font='Monospace',fontSize=10
).encode(x=X_axis,y=Y_axis,text='Display')
#Plotting the Points on the Line on the chart
Temperature_Graph2=alt.Chart(Temperature).mark_circle(color='red').encode(x=X_axis,y=Y_axis,text='Display')
```

```python
Final_Temp=Chart_Temp+Temperature_Graph+Temperature_Graph2
Final_Temp = Final_Temp.properties(height=200,width=1200)
```

```python
Final_Temp
```

```python
#Creating the Survivor Chart
army_chart = alt.Chart(Army).mark_trail().encode(
    longitude='LONP',latitude='LATP',
    size=alt.Size('SURV',
        scale=alt.Scale(range=[1,60])
        ),
    detail='DIV',
    color=alt.Color(
        'DIR',
        scale=alt.Scale(
            domain=['A', 'R'],
            range=[red, green]
        )),
).project(type="mercator")
```

```python
Final_Army = army_chart.properties(height=300,width=1200)
```
